### Adversarial SuperstarGAN: Performance Evaluation & Metrics

```python
import json
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
arguments_log = "logs/arguments_log.json"

with open(arguments_log, "r") as file:
    logs = json.load(file)

df_arguments = pd.DataFrame(logs)

# Move "id" to the first column index.
if "id" in df_arguments.columns:
    cols = ["id"] + [col for col in df_arguments.columns if col != "id"]
    df_arguments = df_arguments[cols]

display(df_arguments)
```

```python
df_arguments.to_csv("logs/arguments_log.csv", index=False)
print("[ INFO ] CSV file saved successfully.")
```

```python
classification_log = "logs/classification_log.json"

with open(classification_log, "r") as file:
    logs = json.load(file)

df_classification = pd.DataFrame(logs)

# Move "id" to the first column index.
if "id" in df_classification.columns:
    cols = ["id"] + [col for col in df_classification.columns if col != "id"]
    df_classification = df_classification[cols]
    
df_classification.head()
```

```python
# Get the "id" with the highest the "adversarial_accuracy"
df_classification[df_classification['adversarial_accuracy']==df_classification['adversarial_accuracy'].max()]
```

```python
df_classification.to_csv("logs/classification_log.csv", index=False)
print("[ INFO ] CSV file saved successfully.")
```
